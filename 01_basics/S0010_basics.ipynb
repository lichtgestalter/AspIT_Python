##### Arithmetic operators

```python
a = 5
b = 3
print(a + b)  # addition
```

```python
c = a - b  # subtraction
print(c * c)  #  multiplication
```

```python
print(a % b)  # remainder (sometimes called modulo)
```

```python
print(c)
```

```python
print(c % c)
```

```python
print(a ** c)  # exponentiation
```

```python
print(a / c)  # division
print(type(a / c))
```

```python
print(a // c)  # floor division
print(type(a // c))
```

```python
d = 17
d += 5
print(d)
d -= 18
print(d)
```

##### The in operator

```python
numbers = [3, 1, 8]
print("numbers has type", type(numbers))
print(2 in numbers)
```

```python
print(3 in numbers)
```
